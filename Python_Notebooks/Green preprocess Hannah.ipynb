```python
import pandas as pd
%pylab inline
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("C:/Users/Hannah/Desktop/UT Austin/Analytics - MSBA/summer 19/predictive modeling/STA380-1-master/STA380-1-master/data/greenbuildings.csv")
df.head(5)
```

```python
len(df)
```

```python
#create column merging class a and b
for i in range(len(df['class_a'])):
    if df.loc[i, 'class_a'] == 1:
        df.loc[i, 'class'] = 'a'
        
    if df.loc[i, 'class_b'] == 1:
        df.loc[i, 'class'] = 'b'
        
    if df.loc[i, 'class_a'] != 1 and df.loc[i, 'class_b'] != 1:
        df.loc[i, 'class'] = 'c'
```

```python
# Preprocessing for plotting mean rent of buildings per green status by class
df2 = df.groupby(['class','green_rating'])[['Rent']].agg('mean').reset_index()
df2
```
